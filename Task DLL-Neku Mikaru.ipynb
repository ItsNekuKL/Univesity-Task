# Latihan 1 - Input Double LinkedList

Diberikan sebuah linked list ganda dengan n elemen. Tugasnya adalah mengembalikan linked list ganda tersebut sebagai sebuah array.

Catatan: Kode driver akan mencetak daftar maju dan mundur.

Latihan:

Input: LinkedList: 34<->22

Output:
 [34, 22]
[22, 34]


```python
class Node:
  def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

def toArray(head):
    """
    Mengonversi Doubly Linked List menjadi Array (List Python).
    """
    result = []
    current = head

    # Melakukan traversal dari depan ke belakang
    while current:
        result.append(current.data)
        current = current.next

    return result

# --- Simulasi Kode Driver ---
def print_doubly_linked_list(head):
    forward_list = toArray(head)

    # Mencetak array hasil penelusuran maju
    print(forward_list)

    # Mencetak array hasil penelusuran mundur
    # Karena ini Python, kita bisa menggunakan slicing [::-1]
    # atau membalikkan array secara manual untuk simulasi output driver
    print(forward_list[::-1])

# Contoh Penggunaan:
# Membuat list: 1 <-> 2 <-> 3 <-> 4 <-> 5
n1 = Node(34)
n2 = Node(22)

n1.next = n2; n2.prev = n1

print_doubly_linked_list(n1)
```

# Latihan 2 - PDDBG
Masukan: 5 <-> 15 <-> 25 <-> 35 <-> 45 <-> 55

Keluaran: [5, 15, 25, 35, 45, 55][55, 45, 35, 25, 15, 5]


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

def solve():
    user_input = input("Masukkan elemen (Add 'Space' for next element): ")
    values = list(map(int, user_input.split()))

    if not values:
        print("List kosong")
        return

    head = Node(values[0])
    curr = head
    for i in range(1, len(values)):
        new_node = Node(values[i])
        curr.next = new_node
        new_node.prev = curr
        curr = new_node

    forward_res = []
    temp = head
    while temp:
        forward_res.append(temp.data)
        temp = temp.next

    print(forward_res)          # Forward
    print(forward_res[::-1])    # Backward

solve()
```

# Latihan 4 - Forward Tranversal
> Dalam penelusuran maju (Forward Traversal), kita mulai dari node pertama, yaitu head dari Doubly Linked List, dan terus mengunjungi node berikutnya menggunakan pointer next dari setiap node hingga kita mencapai node terakhir dari linked list.

**Latihan 4**

Diberikan sebuah Doubly Linked List yang berisi data berat badan (dalam kg). Tugasmu adalah menelusuri list tersebut dari kepala (head) hingga akhir, lalu masukkan nilainya ke dalam sebuah array. Jika berat badan lebih dari 60 kg, tambahkan keterangan "Overweight" di samping angka tersebut dalam bentuk string.

Input: 45 <-> 70 <-> 55 <-> 82

Target Output: ['45', '70 Overweight', '55', '82 Overweight']

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

def analisis_berat(head):
    hasil = []
    curr = head # << (A)

    while curr: # << (B)
        if curr.data > 60:
            hasil.append(f"{curr.data} Overweight")
        else:
            hasil.append(str(curr.data))

        curr = curr.next # << (C)
    return hasil

# --- Bagian Input ---
user_input = input("Masukkan berat badan ('Space' for different element)(KG): ")
data_list = list(map(int, user_input.split()))

if data_list:
    head = Node(data_list[0])
    temp = head
    for val in data_list[1:]:
        new_node = Node(val)
        temp.next = new_node # << (D)
        new_node.prev = temp # << (E)
        temp = new_node

    print("Hasil Analisis:", analisis_berat(head))
```

# Latihan 5 - Backward traversal

> Dalam penelusuran mundur (Backward Traversal), kita mulai dari node terakhir, yaitu ekor dari daftar berantai ganda (Doubly Linked List) dan terus mengunjungi node-node sebelumnya menggunakan pointer prev dari setiap node hingga kita mencapai node pertama dari daftar berantai tersebut.


**LATIHAN 5**

Diberikan sebuah Doubly Linked List yang menyimpan urutan langkah dalam sebuah tugas. Kamu diminta untuk mengembalikan urutan langkah tersebut secara terbalik (dari ekor/tail ke kepala/head) untuk melakukan proses "Undo".

Input: Membuka File <-> Mengetik <-> Menyimpan <-> Menutup

Contoh: --- Simulasi Sistem Undo (Doubly Linked List) --- Masukkan langkah-langkah tugas (pisahkan dengan tanda koma atau spasi):

Target Output: Urutan Langkah Asli (Maju): ['File', 'Mengetik', 'Menyimpan', 'Menutup']

Urutan Proses Undo (Mundur): ['Menutup', 'Menyimpan', 'Mengetik', 'File']

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

def sistem_undo():
    print("--- Simulasi Sistem Undo (Doubly Linked List) ---")
    user_input = input("Masukkan langkah-langkah tugas ('Space' For different element): ")
    raw_steps = user_input.replace(',', ' ').split()

    if not raw_steps:
        print("Tidak ada langkah yang dimasukkan.")
        return

    head = Node(raw_steps[0])
    curr = head
    for step in raw_steps[1:]:
        new_node = Node(step)
        curr.next = new_node # (Tugas A: Pointer next menunjuk ke depan)
        new_node.prev = curr # (Tugas A: Pointer prev menunjuk ke belakang)
        curr = new_node      # (Tugas B: Geser current ke node baru)

    tail = head
    while tail.next is not None: # (Tugas C: Berhenti jika next adalah None)
        tail = tail.next         # (Tugas D: Geser tail ke depan)

    undo_result = []
    temp = tail

    while temp is not None: # (Tugas E: Berhenti setelah melewati head)
        undo_result.append(temp.data) # (Tugas F: Simpan data ke array)
        temp = temp.prev # (Tugas G: Geser pointer ke belakang)

    print("\nUrutan Langkah Asli (Forward):")
    print(raw_steps)
    print("\nUrutan Proses Undo (Backward):")
    print(undo_result)

if __name__ == "__main__":
    sistem_undo()
```

# Latihan 6 - Sisip Node di depan



Sebuah bank menggunakan sistem Doubly Linked List untuk mengelola antrean nasabah di layar monitor. Saat ini, sudah ada 5 nasabah yang mengantre secara berurutan. Tiba-tiba, seorang Nasabah VIP datang. Sesuai kebijakan bank, nasabah VIP harus langsung diletakkan di posisi paling depan (nomor antrean 1) tanpa menggeser nasabah lain keluar dari list.

Sebagai pengembang sistem, tugasmu adalah melengkapi fungsi untuk menyisipkan nasabah VIP tersebut ke awal antrean.

Tugas teman-teman:
Lengkapi bagian # ___ pada kode di bawah ini agar sistem dapat:

1. Menerima 5 ID nasabah awal.
2. Membangun koneksi dua arah (next & prev) antar nasabah.
3. Menyisipkan 1 ID nasabah VIP di posisi paling depan.

```python
class Nasabah:
    def __init__(self, id_nasabah):
        self.id_nasabah = id_nasabah
        self.next = None
        self.prev = None

def sisipkan_vip(head, id_vip):
    node_vip = Nasabah(id_vip)# << (A)
    node_vip.next = head # << (B)
    node_vip.prev = None

    if head is not None:
        head.prev = node_vip # << (C)
    return node_vip

# --- PROGRAM UTAMA ---

print("=== Pendaftaran 5 Antrean Awal ===")
# Contoh input: 1001 1002 1003 1004 1005
input_user = input("Masukkan 5 ID Nasabah (pisah spasi): ").split()
ids = [int(x) for x in input_user]

head = None
if len(ids) > 0:
    head = Nasabah(ids[0])
    curr = head

    # Loop - untuk membangun sisa antrean
    for i in range(1, len(ids)):
        baru = Nasabah(ids[i])
        curr.next = baru
        baru.prev = curr # << (D)
        curr = baru

print("\nAntrean Saat Ini:")
t = head
while t:
    print(f"[ID:{t.id_nasabah}]", end=" <-> " if t.next else "\n")
    t = t.next

# --- PROSES VIP ---
print("\n=== Kedatangan Nasabah VIP ===")
vip_id = int(input("Masukkan ID Nasabah VIP: "))

head = sisipkan_vip(head, vip_id)

print("\nAntrean Terbaru (VIP di Depan):")
t = head
while t:
    print(f"[ID:{t.id_nasabah}]", end=" <-> " if t.next else "")
    t = t.next # Menambahkan ini agar tidak terjadi infinite loop
```

# Latihan 7 - Sisipkan Node Di belakang

Sebuah kereta logistik direpresentasikan sebagai Doubly Linked List. Setiap gerbong (node) memiliki nomor ID unik. Masalahnya, gerbong baru selalu harus dipasang di urutan paling belakang agar tidak mengganggu susunan gerbong yang sudah ada di depan.

Tugasmu adalah melengkapi fungsi untuk mencari gerbong terakhir, lalu menyambungkan gerbong baru tersebut sehingga menjadi ekor (tail) yang baru.

**Lengkapi bagian # ___ pada kode di bawah ini agar sistem dapat menerima 5 ID gerbong awal, lalu menambahkan 1 gerbong tambahan di akhir rangkaian.**

```python
class Gerbong:
    def __init__(self, id_gerbong):
        self.id_gerbong = id_gerbong
        self.next = None
        self.prev = None

def tambah_gerbong_di_akhir(head, id_baru):
    node_baru = Gerbong(id_baru)  # << (A)
    if head is None:
        return node_baru
    curr = head
    while curr.next is not None:  # << (B)
        curr = curr.next          # << (C)
    curr.next = node_baru         # << (D)
    node_baru.prev = curr         # << (E)
    return head

# --- PROGRAM UTAMA ---
print("=== Pendaftaran 5 Gerbong Awal ===")
input_user = input("Masukkan 5 ID Gerbong (pisah spasi): ").split()
ids = [int(x) for x in input_user]

head = None
if len(ids) > 0:
    head = Gerbong(ids[0])
    temp = head
    for i in range(1, len(ids)):
        baru = Gerbong(ids[i])
        temp.next = baru
        baru.prev = temp
        temp = baru

print("\nSusunan Kereta Saat Ini:")
t = head
while t:
    print(f"[ID:{t.id_gerbong}]", end=" <-> " if t.next else "\n")
    t = t.next # fungsi ini agar tidak infinite loop

# --- PROSES TAMBAH GERBONG ---
print("\n=== Penambahan Gerbong Baru di Akhir ===")
id_akhir = int(input("Masukkan ID Gerbong Tambahan: "))

# Memanggil fungsi penyisipan
head = tambah_gerbong_di_akhir(head, id_akhir)

print("\nSusunan Kereta Terbaru:")
t = head
while t:
    print(f"[ID:{t.id_gerbong}]", end=" <-> " if t.next else "")
    t = t.next # fungsi ini agar tidak infinite loop
```

# Latihan 8 - insert a node after node-2

Diberikan kepala dari sebuah daftar berantai ganda, sebuah posisi p, dan sebuah bilangan bulat x. Tambahkan sebuah node baru dengan nilai x pada posisi tepat setelah node ke- p dalam daftar berantai ganda tersebut dan kembalikan kepala dari daftar yang telah diperbarui.

Catatan: Posisi diindeks dari 0.

image.png

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

def addNode(head, p, x):
    """
    Menambahkan node baru dengan nilai x tepat setelah node ke-p.
    Indeks dimulai dari 0.
    """
    new_node = Node(x)

    if head is None:
        return new_node

    curr = head
    # Langkah 1: Cari posisi node ke-p
    count = 0
    while curr is not None and count < p:
        curr = curr.next
        count += 1

    # Jika posisi p ditemukan
    if curr is not None:
        # Simpan referensi ke node setelah curr
        next_node = curr.next
        # Langkah 2: Hubungkan node baru dengan curr
        curr.next = new_node
        new_node.prev = curr
        # Langkah 3: Hubungkan node baru dengan node selanjutnya (jika ada)
        new_node.next = next_node
        if next_node is not None:
            next_node.prev = new_node

    return head

def print_list(head):
    res = []
    curr = head
    while curr:
        res.append(curr.data)
        curr = curr.next
    print(res)

# --- Contoh Penggunaan ---
# Membuat list awal: 2 - 4 - 5
head = Node(2)
n2 = Node(4)
n3 = Node(5)
head.next = n2; n2.prev = head
n2.next = n3; n3.prev = n2

print("List Awal:")
print_list(head)


p = 2
x = 6 # menambah 6 setelah angka 5
head = addNode(head, p, x)

print(f"List setelah tambah {x} setelah posisi {p}:")
print_list(head)
```

# Latihan 9 - insert a node after node-0

Catatan: Posisi diindeks dari 0

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

def addNode(head, p, x):
    new_node = Node(x)

    if head is None:
        return new_node

    curr = head
    # Cari posisi node ke-p (p=0 berarti berhenti di head)
    count = 0
    while curr is not None and count < p:
        curr = curr.next
        count += 1
    if curr is not None:
        next_node = curr.next

        # Hubungkan node baru ke node saat ini (curr)
        new_node.prev = curr
        new_node.next = next_node

        # Hubungkan node saat ini ke node baru
        curr.next = new_node

        # Jika ada node setelah node baru, hubungkan kembali ke node baru
        if next_node is not None:
            next_node.prev = new_node

    return head

def print_list(head):
    res = []
    curr = head
    while curr:
        res.append(curr.data)
        curr = curr.next
    print(res)

# --- Implementasi ---
# List Awal: 1 - 2 - 3 - 4
head = Node(1)
n2 = Node(2)
n3 = Node(3)
n4 = Node(4)

head.next = n2; n2.prev = head
n2.next = n3; n3.prev = n2
n3.next = n4; n4.prev = n3

print("List Awal:")
print_list(head)

# Task: Tambahkan 5 setelah posisi p=0
p = 0
x = 5
head = addNode(head, p, x)

print(f"\nList setelah tambah {x} setelah posisi {p}:")
print_list(head)
```
